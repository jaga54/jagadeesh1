```python
p25=data["CGPA"].quantile(0.25)
p75=data["CGPA"].quantile(0.75)
iqr=p75-p25
```

```python
print(p25)
print(p75)
print(iqr)
```

```python
lowerrange=p25-(1.5*iqr)
upperrange=p75+(1.5*iqr)
```

```python
print(lowerrange)
print(upperrange)
```

```python
data["CGPA"][data["CGPA"]<lowerrange]=np.nan
data["CGPA"][data["CGPA"]>upperrange]=np.nan
```
